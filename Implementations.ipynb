## 1. KNN Regression and Classification

```python
import numpy as np
import matplotlib.pyplot as plt
import KNN
```

```python
# Creating data
mean = (0, 0)
cov = [[2, 0], [0, 2]]

train_x = np.random.multivariate_normal(mean, cov, 300)
train_y = np.ones((300,1))         
for i,x in enumerate(train_x):
    if np.dot(x, x) > 1:
        train_y[i] = 0
```

```python
knn = KNN.KNN(train_x, train_y, 4, 2, "classification")
knn.predict()

fig = plt.figure(figsize=(10,10))
data = fig.add_subplot(121, aspect='equal', title='Data', xlabel="x_1", ylabel="y_1" )
test = fig.add_subplot(122, aspect='equal', title='Testing using implementation',  xlabel="x_1", ylabel="y_1" )
data.scatter(train_x[:,0], train_x[:,1], c=np.reshape(train_y, (300)))
test.scatter(train_x[:,0], train_x[:,1], c=np.reshape(np.round(knn.predictions), (300)))
```

## 2. Ridge Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import RidgeRegression
```

```python
def error(self, string):
        
    if string == "RMS":
        return np.asscalar(np.mean((self.Y - self.predictions)**2, axis=0))
    elif string == "R2":
        RSS = np.sum((self.Y - self.predictions)**2, axis=0) 
        TSS = np.sum((self.Y - np.mean(self.Y, axis=0))**2, axis=0)
        return np.asscalar(1 - RSS/TSS)
    elif string == "Classification":
        return  (1 - np.mean(self.predictions - self.Y))
    else:
        print("Please choose from RMS, R2 and Classification.") 
```

```python
# Create noisy data
N = 100
dim = 1
slope = 2
noise_sd = 0.3
train_x = np.random.rand(N, dim)*10
train_y = np.sum(train_x * slope , axis=1, keepdims=True) + np.reshape(np.random.normal(0, noise_sd, N), (N, 1))
```

```python
rr = RidgeRegression.RidgeReg(train_x, train_y, 0.5)
```

#### Checking with scikit implementation

```python
from sklearn.linear_model import Ridge
sk_rr = Ridge(alpha=0.5)
sk_rr.fit(train_x, train_y)
sk_rr.score(train_x, train_y)
```

## 3. Lasso Regression using Coordinate descent

```python
import numpy as np
import LassoRegression
```

```python
# Create data where most of the predictors are useless (99/100 are random).

N = 1000
dim = 100
slope = 2
noise_sd = 0.3
train_x = np.random.rand(N, dim)
train_y = np.reshape(train_x[:,0] * slope, (N, 1))
print(train_x.shape, train_y.shape)
```

```python
lasso = LassoRegression.LassoReg(train_x, train_y, 1)
lasso.predict()
print("Parameters:", "\n", np.round(lasso.W.T))
```

```python
from sklearn.linear_model import Lasso
lass_sk = Lasso(0)
lass_sk.fit(train_x, train_y)
lass_sk.score(train_x, train_y)
```

## 4. Regerssion with Fourier Basis - 1D

```python
import numpy as np
import matplotlib.pyplot as plt
import FourierRegression1D
    
```

```python
# Create noisy data 
N = 100
dim = 1
noise_sd = 0.4
train_x = np.reshape(np.sort(np.random.uniform(0, 4, N)), (N, dim))
train_y = np.reshape(np.exp(train_x), (N, 1)) + np.reshape(np.random.normal(0, noise_sd, N), (N, 1))
print(train_x.shape, train_y.shape)
```

```python
freg = FourierRegression1D.FourierReg(train_x, train_y, k=4)
freg.fourier_reg()
freg.predict()
plt.scatter(train_x, train_y)
plt.plot(train_x, freg.predictions, color="red", label="fourier")
```

#### Comparing the performance of the above three methods for noisy, exponential data

```python
import matplotlib.pyplot as plt

rr = RidgeRegression.RidgeReg(train_x, train_y, 0)
rr.predict()

lasso = LassoRegression.LassoReg(train_x, train_y, 1)
lasso.predict()

plt.xlabel("x")
plt.ylabel("y")
plt.title("Comparing the performance of lasso, ridge and fourier for noisy exponential data")
#plt.ylim([0, 20])
plt.plot(train_x, rr.predictions, color="orange", label="ridge")
plt.plot(train_x, lasso.predictions, color="purple", label="lasso")
plt.plot(train_x, freg.predictions, color="red", label="fourier")
plt.scatter(train_x, train_y)
plt.legend()
plt.show()
```

## 5.  Logistic Regression (Binary)

```python
import numpy as np
import matplotlib.pyplot as plt
import LogisticRegressionBinary
```

```python
# Create data
N = 100
p = 2
train_x = np.random.uniform(-4, 4, (N, p))
train_y = np.ones((N, 1))
for i,x in enumerate(train_x):
    if x[0] < 0:
        train_y[i] = 0 
```

```python
logreg = LogisticRegressionBinary.LogisticReg(train_x, train_y, 0.01, 10)
logreg.predict()

fig = plt.figure(figsize=(10,10))
data = fig.add_subplot(121, aspect='equal', title='Data', xlabel="x_1", ylabel="y_1" )
test = fig.add_subplot(122, aspect='equal', title='Testing using implementation',  xlabel="x_1", ylabel="y_1" )
data.scatter(train_x[:,0], train_x[:,1], c=np.reshape(train_y, (N)))
test.scatter(train_x[:,0], train_x[:,1], c=np.reshape(np.round(logreg.predictions), (N)))
```

#### Checking with scikit imlementation

## 6. Linear Discriminant Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import LinearDiscriminantAnalysis
```

```python
# Create data

mean1 = (-2, -4)
mean2 = (-1, 2)
mean3 = (4, 0)
cov = [[2, 1], [1, 2]]

data1 = np.random.multivariate_normal(mean1, cov, 1000)
data2 = np.random.multivariate_normal(mean2, cov, 1000)
data3 = np.random.multivariate_normal(mean3, cov, 1000)

train_x = np.concatenate((data1, data2, data3), axis=0)
train_y = np.concatenate((np.ones(1000), np.zeros(1000), np.ones(1000)*-1), axis=0)

```

```python
lda = LinearDiscriminantAnalysis.LDA(train_x, train_y)
lda.predict()
lda.predictions

fig = plt.figure(figsize=(10,10))
data = fig.add_subplot(121, aspect='equal', title='Data', xlabel="x_1", ylabel="y_1" )
test = fig.add_subplot(122, aspect='equal', title='Testing using implementation',  xlabel="x_1", ylabel="y_1" )
data.scatter(train_x[:,0], train_x[:,1], c=np.reshape(train_y, (3000)))
test.scatter(train_x[:,0], train_x[:,1], c=np.reshape(np.round(lda.predictions), (3000)))

 
```

## 7. Support Vector Machine using SMO

```python
# Support Vector Machine using simplified sequential minimal optimization (SMO).
# This code is based on the notes on Support Vector Machines from Standford's CS229.

import numpy as np
import matplotlib.pyplot as plt
import SupportVectorMachineBinary

```

```python
train_x = np.random.uniform(-4, 4, (300, 2))
train_y = np.ones((300, 1))
for i, x in enumerate(train_x):
    if x[0] > -2 and x[0] < 2:
        train_y[i] = -1
```

```python
svm = SupportVectorMachineBinary.SVM(train_x, train_y, 1, 0.01, 30, "exp", 3, 1)
svm.predict()

fig = plt.figure(figsize=(10,10))
data = fig.add_subplot(121, aspect='equal', title='Data', xlabel="x_1", ylabel="y_1" )
test = fig.add_subplot(122, aspect='equal', title='Testing using implementation',  xlabel="x_1", ylabel="y_1" )
data.scatter(train_x[:,0], train_x[:,1], c=np.reshape(train_y, (300)))
test.scatter(train_x[:,0], train_x[:,1], c=np.reshape(np.round(svm.predictions), (300)))
```

```python
from sklearn.svm import SVC
svc_sk = SVC(kernel='rbf', degree=3)
svc_sk.fit(train_x, train_y)
plt.axis('equal')    
plt.scatter(train_x[:,0], train_x[:,1], c=svc_sk.predict(train_x)) 
```

## 8.  Decision Tree

```python
import numpy as np
import DecisionTrees
```

```python
# Create data

train_x = np.random.uniform(0,10, (200, 2))
train_y = np.ones((200,1))
for i,x in enumerate(train_x):
    if x[0] >= 5 and x[1] > 5:
        train_y[i] = 10
    elif x[0] >= 5:
        train_y[i] = 5    
```

```python
tree = DecisionTrees.DecisionTree(train_x, train_y, 2, "classification")
tree.predict()

fig = plt.figure(figsize=(10,10))
data = fig.add_subplot(121, aspect='equal', title='Data', xlabel="x_1", ylabel="y_1" )
test = fig.add_subplot(122, aspect='equal', title='Testing using implementation',  xlabel="x_1", ylabel="y_1" )
data.scatter(train_x[:,0], train_x[:,1], c=np.reshape(train_y, (200)))
test.scatter(train_x[:,0], train_x[:,1], c=np.reshape(np.round(tree.predictions), (200)))
```
